```python
%matplotlib notebook 

#manipular el plot

import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

#: metodo de ambito , todo lo indexado pertenece a la funcion que se cree 
# atributos, variables que se guardan en la memoria del computador 
# set usar atributos fuera de la clase 
# get obtener atributos fuera de la clase
# :: scale, salto en una lista
```

```python
A=1
B=2
omega= 2*np.pi/10
N=100
```

```python
t= np.linspace(0,10,N)
t
r=np.zeros((N,3))
```

```python
r[:,0]=A*np.cos(omega*t)
r[:,1]=A*np.sin(omega*t)
r[:,2]=B*t
```

```python
fig= plt.figure(figsize=(5,5))
ax = fig.add_subplot(111,projection="3d")
ax.scatter(r[:,0],r[:,1],r[:,2],color='k', marker='o',s=50)
ax.view_init()
```

```python
#campos vectoriales
n=8
#mallado
x=np.linspace(0,1,n)
y=np.linspace(0,1,n)
X,Y = np.meshgrid(x,y)
#meshgrid 
fig= plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.scatter(X,Y)
ax1 = fig.add_subplot(122)
#guardar 
Ex= np.zeros((n,n))
Ey=np.zeros((n,n))
for i in range(n):
    for j in range(n):
        Ex[i,j]=1
        Ey[i,j]=-2.

for i in range(n):
    for j in range(n):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j])
        
ax1.streamplot(X,Y,Ex.T,Ey.T,density=2)
```
